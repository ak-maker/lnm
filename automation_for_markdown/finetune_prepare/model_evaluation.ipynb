```python
model_responses = []
instruction_and_answer_pairs = []

for i, entry in enumerate(test_descriptions_list):
    print(entry)
    response = client.chat.completions.create(
        model="gpt-35-turbo-ft-jiewen",
        messages=[
            {"role": "system", "content": "You are an expert in Sionna, capable of writing code and illustrating information like formulas and APIs. Your proficiency in Sionna allows you to simulate and analyze various aspects of wireless communication systems effectively."},
            {"role": "user", "content": f"Answer the following question about Sionna: {entry}. Remember to return your answer with markdown format. Whenever an answer requires code, it is imperative to provide the complete code without fail. Under no circumstances should there be any laziness or shortcuts in providing the necessary code"},
        ]
    )
    response_lines = response.choices[0].message.content.strip().split('\n')
    instruction_and_answer_pairs.append((entry, response_lines))


with open('E:\\python\\lnm\\automation_for_markdown\\finetune_prepare\\3_11\\sample_code.md', 'w', encoding='utf-8') as file:
    for instruction, answer in instruction_and_answer_pairs:
        file.write(f"Instruction: {instruction}")
        answer = '\n'.join(answer)
        file.write(f"Answer: {answer}")
        file.write("\n\n")
```
